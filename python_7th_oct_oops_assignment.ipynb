```python
# problem 1:
class bankaccount:
             
    def __init__(self, account_holder_name, account_number, balance = 0):
        self.account_number = account_number
        self.account_holder_name = account_holder_name
        self.balance = balance
        
    def deposit(self, amount):
        self.amount = amount
        if amount > 0:
            self.balance += amount
            print(f"deposited amount is {self.amount} and balance is {self.balance}")
        else:
            print("Deposited amount should be greater than zero")
            
    def withdraw(self, amount_to_be_withdrawed):
        self.w = amount_to_be_withdrawed
        if self.balance >= self.w:
            self.balance -= self.w
            print(f"{self.account_holder_name} withdrawed money{self.w} and remaining balance is {self.balance}")
        elif(self.balance < self.w):
            print("Sorry, not enough balance in your bankaccount")
   
    def print_balance(self):
        print(f"{self.account_holder_name} and {self.account_number} has current balance {self.balance}")
             
dishant = bankaccount("dishant gupta", "1234567")
dishant.deposit(10000)
dishant.withdraw(3500)
dishant.print_balance()
```

```python
# Problem 3: Vehicle Rental Create a class representing a vehicle rental system. 
#          Implement methods to rent a vehicle, return a vehicle, and display available vehicles.
import logging

class Vehicle:
    def __init__(self, vehicle_type, available=True):
        self.vehicle_type = vehicle_type
        self.available = available

class VehicleRentalSystem:
    def __init__(self):
        self.vehicles = []
        self.logger = logging.getLogger('vehicle_rental_system')
        self.logger.setLevel(logging.DEBUG)
        self.log_formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')
        self.file_handler = logging.FileHandler('vehicle_rental_log.log')
        self.file_handler.setLevel(logging.DEBUG)
        self.file_handler.setFormatter(self.log_formatter)
        self.logger.addHandler(self.file_handler)

    def rent_vehicle(self, vehicle_type):
        for vehicle in self.vehicles:
            if vehicle.vehicle_type == vehicle_type and vehicle.available:
                vehicle.available = False
                self.logger.info(f"{vehicle_type} rented successfully!")
                return
        self.logger.warning(f"Sorry, {vehicle_type} is not available for rent.")

    def return_vehicle(self, vehicle_type):
        for vehicle in self.vehicles:
            if vehicle.vehicle_type == vehicle_type and not vehicle.available:
                vehicle.available = True
                self.logger.info(f"{vehicle_type} returned successfully!")
                return
        self.logger.warning(f"Invalid vehicle type or it's already returned.")

    def display_available_vehicles(self):
        available_vehicles = [vehicle.vehicle_type for vehicle in self.vehicles if vehicle.available]
        if available_vehicles:
            self.logger.info("Available vehicles:")
            for vehicle_type in available_vehicles:
                self.logger.info(vehicle_type)
        else:
            self.logger.warning("No vehicles available for rent.")

# Example usage:
system = VehicleRentalSystem()
system.vehicles.append(Vehicle("Car"))
system.vehicles.append(Vehicle("Bike"))
system.vehicles.append(Vehicle("Truck"))

system.display_available_vehicles()  # Display available vehicles
system.rent_vehicle("Car")  # Rent a car
system.display_available_vehicles()  # Display available vehicles after renting
system.return_vehicle("Car")  # Return the car
system.display_available_vehicles()  # Display available vehicles after returning
```

```python
#Problem 4: Library Catalog Create classes representing a library and a book. 
#          Implement methods to add books to the library, borrow books, and display available books.
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.is_borrowed = False

    def __str__(self):
        return f"{self.title} by {self.author}"

    def borrow(self):
        if not self.is_borrowed:
            self.is_borrowed = True
            return True
        else:
            return False

    def return_book(self):
        self.is_borrowed = False


class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def borrow_book(self, title):
        for book in self.books:
            if book.title == title and not book.is_borrowed:
                book.borrow()
                return f"You have borrowed '{title}'. Enjoy reading!"
        return "Book not available for borrowing."

    def return_book(self, title):
        for book in self.books:
            if book.title == title and book.is_borrowed:
                book.return_book()
                return f"Thank you for returning '{title}'."
        return "Book not found or already returned."

    def display_available_books(self):
        available_books = [book for book in self.books if not book.is_borrowed]
        if available_books:
            print("Available books:")
            for book in available_books:
                print(book)
        else:
            print("No books available at the moment.")
```

```python
# problem 5: Product Inventory Create classes representing a product and an inventory system. 
# Implement methods to add products to the inventory, update product quantity, and display available products.
class Product:
    def __init__(self, product_id, name, price, quantity):
        self.product_id = product_id
        self.name = name
        self.price = price
        self.quantity = quantity

class Inventory:
    def __init__(self):
        self.products = []

    def add_product(self, product):
        self.products.append(product)
        print(f"Product '{product.name}' added to inventory.")

    def update_quantity(self, product_id, new_quantity):
        for product in self.products:
            if product.product_id == product_id:
                product.quantity = new_quantity
                print(f"Quantity updated for product '{product.name}'.")

    def display_available_products(self):
        if not self.products:
            print("Inventory is empty.")
        else:
            print("Available products:")
            for product in self.products:
                print(f"ID: {product.product_id} - {product.name} - Quantity: {product.quantity}")

# Example usage:
inventory = Inventory()

# Adding products to inventory
product1 = Product(1, "Keyboard", 20.0, 50)
product2 = Product(2, "Mouse", 10.0, 80)

inventory.add_product(product1)
inventory.add_product(product2)

# Display available products
inventory.display_available_products()

# Update quantity of a product
inventory.update_quantity(1, 70)

# Display available products after updating quantity
inventory.display_available_products()
```

```python
#Problem 6: Shape Calculation Create a class representing a shape with attributes like length, width, and height.
#Implement methods to calculate the area and perimeter of the shape.
import logging as log
log.basicConfig(filename = "test.log", level = log.DEBUG)
class shape:
    def __init__(self, length, breath, height):
        self.l = length
        self.b = breath
        self.h = height
    
    def cal_area(self):
        area = 2 *((self.l * self.b) + (self.b * self.h) + (self.h * self.l))
        self.area = area
        log.info("Calculating area:- ")
        log.info(f"Area of shape of dimensions {self.l}X{self.b}X{self.h} is :- {self.area}")
    
    def cal_perimeter(self):
        self.perimeter = 4*(self.l + self.b + self.h)
        log.info("Calculating perimeter:-")
        log.info(f"Perimeter of shape of dimensions {self.l}X{self.b}X{self.h} is : {self.perimeter}")
s1 = shape(2, 3, 4)
s1.cal_area()
s1.cal_perimeter()
```

```python
#Problem 9: Social Media Profile Create a class representing a social media profile with attributes like username and posts.
#           Implement methods to add posts, display posts, and search for posts by keyword.
class SocialMediaProfile:
    def __init__(self, username):
        self.username = username
        self.posts = []

    def add_post(self, content):
        self.posts.append(content)

    def display_posts(self):
        if not self.posts:
            print("No posts to display.")
        else:
            print(f"Posts by {self.username}:")
            for index, post in enumerate(self.posts, start=1):
                print(f"Post {index}: {post}")

    def search_posts(self, keyword):
        found_posts = [post for post in self.posts if keyword in post]
        if found_posts:
            print(f"Posts containing '{keyword}' by {self.username}:")
            for idx, post in enumerate(found_posts, start=1):
                print(f"Post {idx}: {post}")
        else:
            print(f"No posts containing '{keyword}' found.")

# Example usage:
profile = SocialMediaProfile("user123")
profile.add_post("Hello, this is my first post.")
profile.add_post("Enjoying the weekend!")
profile.add_post("Exploring new places.")

profile.display_posts()

profile.search_posts("weekend")
```

```python
#Problem 10: ToDo List Create a class representing a ToDo list with attributes like tasks and due dates 
#            Implement methods to add tasks, mark tasks as completed, and display pending tasks.
import logging as log
log.basicConfig(filename = "Todo.log", level = log.DEBUG)
class ToDoList:
    def __init__(self):
        self.tasks = {}

    def add_task(self, task, due_date):
        self.tasks[task] = due_date

    def mark_completed(self, task):
        if task in self.tasks:
            del self.tasks[task]
            log.info(f"Task '{task}' marked as completed.")
        else:
            log.info(f"Task '{task}' not found in the list.")

    def display_pending_tasks(self):
        if not self.tasks:
            log.info("No pending tasks.")
        else:
            log.info("Pending Tasks:")
            for task, due_date in self.tasks.items():
                log.info(f"Task: {task} - Due Date: {due_date}")

# Example Usage
todo = ToDoList()

# Adding tasks
todo.add_task("Complete report", "2023-12-20")
todo.add_task("Buy groceries", "2023-12-18")
todo.add_task("Call John", "2023-12-22")

# Display pending tasks
todo.display_pending_tasks()

# Marking a task as completed
todo.mark_completed("Buy groceries")

# Display pending tasks after marking a task as completed
todo.display_pending_tasks()
```
